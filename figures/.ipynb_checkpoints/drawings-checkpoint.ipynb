```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('ggplot')
%config InlineBackend.figure_format = 'svg'

colors = plt.rcParams['axes.prop_cycle'].by_key()['color']
```

```python
x = [1, 10, 10, 1]
y = [1, 0.75, 1.25, 1]
plt.plot(x, y, '-o')
plt.axis((-1,13,-0.5,2.5))
plt.annotate(r'$a$', xy=(10.5, 1))
plt.annotate(r'$\theta$', xy=(4, 0.95))
plt.annotate(r'$D$', xy=(5.5, 0.3))
plt.annotate('', xy=(1, 0.5), xycoords='data', xytext=(10, 0.5), textcoords='data', arrowprops={'arrowstyle': '<->', 'color': colors[1]})
plt.axis('off')
plt.savefig('problem-set-1-skinny-triangle.pdf', transparent=True, bbox_inches='tight', pad_inches=0)
```
